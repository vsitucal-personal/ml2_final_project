```python
import numpy as np
import matplotlib.pyplot as plt

# Option parameters
strike_price = 150  # Strike price of the call option
premium = 10        # Premium paid for the call option

# Stock price range
stock_prices = np.linspace(100, 200, 100)

# Calculate P&L for the call option
# Profit if stock price is above the strike price, otherwise loss (premium paid)
pnl = np.maximum(stock_prices - strike_price, 0) - premium

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(stock_prices, pnl, label='Call Option P&L', color='b')

# Add labels and title
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Profit and Loss')
plt.title('Profit and Loss of a Call Option')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(strike_price, color='red', linewidth=0.5, linestyle='--', label='Strike Price')
plt.legend()
plt.grid(False)

print(f"Premium: {premium}")
print(f"Strike Price: {strike_price}")

plt.show()
```
